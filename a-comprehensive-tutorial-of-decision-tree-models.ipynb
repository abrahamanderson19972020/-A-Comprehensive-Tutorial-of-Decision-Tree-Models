```python
#First we import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<font color="blue">
A Decision Tree is a supervised algorithm used in machine learning. It is using a binary tree graph (each node has two children) to assign for each data sample a target value. The target values are presented in the tree leaves. To reach to the leaf, the sample is propagated through nodes, starting at the root node. In each node a decision is made, to which descendant node it should go. A decision is made based on the selected sampleâ€™s feature. Decision Tree learning is a process of finding the optimal rules in each internal tree node according to the selected metric.
 

<font color="blue">
There are two types of Decision Tree Models as Classification Tree and Regression Tree.

Classification trees used to classify samples, assign to a limited set of values - classes. In scikit-learn it is DecisionTreeClassifier.
    
Regression trees used to assign samples into numerical values within the range. In scikit-learn it is DecisionTreeRegressor.

# 2. Classsification Problems with Decision Tree Model

Step 1: Importing and Exploring the Data

```python
df1=pd.read_csv("../input/kyphosis/kyphosis.csv") #here I import my data
```

```python
df1.info()#here I get overall information about the data
```

```python

sns.pairplot(data=df1, hue="Kyphosis")
#here I get a mutual visualization of all the features
```

Step 2: Separating Data as Features and Target 

```python
X=df1[["Age","Number","Start"]] #here I assign three columns as features 
X
     
```

```python
y=df1["Kyphosis"] #here I assign Kyphosis column as the target
y
```

Step 3: Splitting Data as Train and Test Set

```python
from sklearn.model_selection import train_test_split #Here I import function from sklearn.model_selection to split data
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2)
#First comes features(X), later target(y),later test_size. 
#Here I assign %20 of data as test size by writing test_size=0.2, %80 will automatically will be the train size
```

```python
X.shape #This is the size of X, all features before separation as train and test subsets
```

```python
X_test.shape#This is the size of X test set, features in the test data subset
```

```python
X_train.shape #This is the size of X train set, features in the train data subset
```

```python
y_train.shape #This is the size of y train set, target column in the train data subset
```

```python
y_test.shape #This is the size of y test set, target column in the test data subset
```

Step 4: Implementing Algorithm

```python
from sklearn.tree import DecisionTreeClassifier 
#I call decision tree classifier because my target column has categorical values as absent and presebt, not continuous
```

```python
dtree=DecisionTreeClassifier() # here we create an instance of the algorithm
```

```python
dtree.fit(X_train,y_train) # here the algorithm fits the train data and creates best model to predict y,target values
```

```python
predictions=dtree.predict(X_test) # here the algorithm makes predictions for real y_test data
#we have to only write here X_test data, y_test data because we do not make the algorithm know what the real values is
```

Step 5: Evaluation the Performance of the Algorithm

```python
from sklearn.metrics import confusion_matrix, classification_report
#we import confusion_matrix and classification_report for performance measurement
```

Confusion matrix) is a table with two rows and two columns that reports the number of false positives, false negatives, true positives, and true negatives. 

```python
print(confusion_matrix(y_test,predictions))
#The intersection between first row and column represents True Positives,in our case true number of present kyphosis cases the algorithm predicted truely
#The intersection between second row and first column represents True Negatives, absence of kyphosis cases the algorithm predicted truely
#The intersection between first row and second column represents False Negatives,the algorithm predicted as absence of kyphosus, but it is labeled present in the real data
#The intersection between first row and second column represents True Negatives,the algorithm predicted as absence of kyphosus,which also match with real data
```

```python
print(classification_report(y_test,predictions)) 
#This is classification report of the predictions of the algorithm compared to the real data
```
